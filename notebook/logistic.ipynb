```python
import numpy as np
import pandas as pd
import sys
from pathlib import Path
```

```python
BASE_DIR = str(Path().cwd().parent.resolve())
sys.path.insert(0, BASE_DIR)
_chained = 'unchained'


DF = pd.read_csv(f'{BASE_DIR}/data/imputed/words/{_chained}/fake_job_postings_rule_added.csv')
DF.head()
```

```python
DF.info()
```

```python
col_list = DF.columns[[7,8,9,16,17,18,19,20,21,22,23]]
tar_col = 'fraudulent'
```

```python
import statsmodels.api as sm

logit = sm.Logit(DF[tar_col], DF[col_list])
result = logit.fit()
result.summary2()
```

귀무가설: 독립변수 하나가 종속변수에 영향을 미치지 않는다

--> p-value가 0.05보다 작으면 귀무가설 기각이므로 유의미하다.

```python
result.pvalues < 0.05
```
